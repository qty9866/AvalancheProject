```python
from netCDF4 import Dataset

# 替换为你的 WRF 输出文件路径
file_path = "/WRF/Product_WRF/WRFOUT/wrfout_d01_2025-04-22_00:00:00"

# 打开 NetCDF 文件
ncfile = Dataset(file_path)

# 查看网格维度信息
print("Domain dimensions:")
print(f"Time steps: {len(ncfile.dimensions['Time'])}")
print(f"west_east (x方向格点数): {len(ncfile.dimensions['west_east'])}")
print(f"south_north (y方向格点数): {len(ncfile.dimensions['south_north'])}")

# 查看经纬度范围（取第一个时间步）
lat = ncfile.variables['XLAT'][0,:,:]  # 纬度 [deg]
lon = ncfile.variables['XLONG'][0,:,:] # 经度 [deg]

print("\n纬度范围:")
print(f"最小纬度: {lat.min():.4f}°N")
print(f"最大纬度: {lat.max():.4f}°N")

print("\n经度范围:")
print(f"最小经度: {lon.min():.4f}°E")
print(f"最大经度: {lon.max():.4f}°E")

# 关闭文件
ncfile.close()
```
